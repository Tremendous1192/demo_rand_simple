# Demonstration of distribution

```python
# ---- ライブラリをimportする ----

# ---- DataFrameライブラリをimportする ----
import polars as pl
import polars.selectors as cs
import numpy as np

#  ---- DataFrameの設定を初期化する ----
pl.Config(fmt_str_lengths = 100, tbl_cols = 100, tbl_rows = 10)
np.set_printoptions(precision = 3, floatmode = 'fixed')

# ---- プロットライブラリをimportする ----
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from matplotlib import gridspec

# ※日本語でプロットしたい場合、英語フォントはコメントアウトする
import japanize_matplotlib
#plt.rcParams['font.family'] = 'Times New Roman'

# ---- プロットの設定を初期化する ----
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams["font.size"] = 15
plt.rcParams['xtick.labelsize'] = 9
plt.rcParams['ytick.labelsize'] = 24
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['axes.grid'] = True
```

```python
name = "Bernoulli"# distribution
df = (
    pl.read_csv("../data/" + name + ".csv")
    # プロット時間を短くするために、値を丸める
    .with_columns([
        pl.col("Standard distribution").round(1),
        pl.col("Custom parameter distribution").round(1),
    ])
    # プロット処理を短くするために、外れ値を削除する
    .unpivot(cs.numeric())
    .filter(
        pl.col("value").abs() < 100
    )
)

# Show first 5 rows data
display(df.head())

# Plot distribution
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (6, 6))
sns.histplot(
    data = df,
    x = "value",
    hue = "variable",
    ax = ax,
)
plt.suptitle(name + " distribution")
plt.tight_layout()
```

```python
name = "LogNormal"# distribution
df = pl.read_csv("../data/" + name + ".csv")

# Show first 5 rows data
display(df.head())

# Plot distribution
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (6, 6))
sns.histplot(
    data = (
        df
        # プロット時間を短くするために、値を丸める
        .with_columns([
            pl.col("Standard distribution").round(1),
            pl.col("Custom parameter distribution").round(1),
        ])
        # パラメータで層別するためにunpivotする
        .unpivot(cs.numeric())
        # プロット処理を短くするために、外れ値を削除する
        .filter(
            pl.col("value").abs() < 100
        )
    ),
    x = "value",
    hue = "variable",
    ax = ax,
)
plt.suptitle(name + " distribution")
plt.setp(ax, xlim = (-2, 10))
plt.tight_layout()
```
